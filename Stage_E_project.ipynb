```python
#importing libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#estabilishing connection with web page address
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"
}
```

```python
#scrapping the data 
data = {"title":[],"review":[], "year":[]}
for x in range(1,76):
  pages = requests.get(f'https://www.amazon.com/s?k=amazon+shopping+online&i=stripbooks&rh=n%3A283155&dc&page={x}&adgrpid=84486094307&gclid=CjwKCAjwwab7BRBAEiwAapqpTDYtTN8aym2q_rlYn1VdmofKw0_bzkd6xjaCl8cMY4syjbPZMKNMjxoCLIYQAvD_BwE&hvadid=393524136919&hvdev=c&hvlocphy=1010294&hvnetw=g&hvqmt=b&hvrand=2170543926145783317&hvtargid=kwd-469356929&hydadcr=22365_10729094&qid=1635883021&rnid=2941120011&ref=sr_pg_{x}',
                        headers=headers)
  soup = BeautifulSoup(pages.content, "html.parser")

  review=soup.find_all('div', {"class": "a-row a-size-small"})
  for i in review:
    for j in i.find_all('span', {"class": "a-size-base"}):
      data["review"].append(j.text)
      


  title = soup.find_all('h2', {"class": "a-size-mini a-spacing-none a-color-base s-line-clamp-2"})
  for titles in title:
    data["title"].append(titles.text)

  for year in soup.find_all('span', { 'class':'a-size-base a-color-secondary a-text-normal' }):
      data["year"].append(year.text)
```

```python
#storing the data in a dataframe
books= pd.DataFrame.from_dict(data, orient='index')   
books.T
```

```python
#saving the stored data into a csv file 
books.T.to_csv('books.csv', index=False) 
```

```python
#dropping missing data
df.dropna(axis=0, inplace= True)
```

```python
#confirming if there are missing values left
df.isna().sum()
```

```python
# extracting only the year values from the year column
df['year'] = df['year'].str[-4:]

#removing "," from the reviews column
df['review'] = df['review'].str.replace(",","")
```

```python
#converting year and review columns to integers
df['year'] = df['year'].astype(int)
df['review'] = df['review'].astype(int)
```

```python
#removing unwanted data(ad that was scrapped with data)
df = df.iloc[1: , :]
```
